```python

import os, cv2, warnings
import numpy as np
import torch
import torch.nn as nn
import torchvision.models as models
import torchvision.transforms as T
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')

CHECKPOINT_DIR = '/content/drive/MyDrive/PBL4/checkpoints_resnet50_4class'
DATA_DIR       = '/content/Dataset'
os.makedirs(CHECKPOINT_DIR, exist_ok=True)

CLASS_MAPPING = {
    'Non_Demented'      : 0,
    'Very_Mild_Demented': 1,
    'Mild_Demented'     : 2,
    'Moderate_Demented' : 3
}
CLASS_NAMES = ['Non', 'Very_Mild', 'Mild', 'Moderate']

def prepare_data(data_dir):
    image_paths, labels = [], []
    for cls_name, label in CLASS_MAPPING.items():
        cls_path = os.path.join(data_dir, cls_name)
        if not os.path.exists(cls_path):
            continue
        for fname in os.listdir(cls_path):
            if fname.endswith(('.jpg','.jpeg','.png')):
                image_paths.append(os.path.join(cls_path, fname))
                labels.append(label)
    labels = np.array(labels)
    print("Phân phối:")
    for i, n in enumerate(CLASS_NAMES):
        print(f"  {n:12s}: {(labels==i).sum()}")
    return image_paths, labels

train_transform = T.Compose([
    T.ToPILImage(),
    T.Resize((224, 224)),
    T.RandomHorizontalFlip(),
    T.RandomRotation(15),
    T.ColorJitter(brightness=0.2, contrast=0.2),
    T.ToTensor(),
    T.Normalize([0.485], [0.229])
])

val_transform = T.Compose([
    T.ToPILImage(),
    T.Resize((224, 224)),
    T.ToTensor(),
    T.Normalize([0.485], [0.229])
])

class AlzheimerDataset(Dataset):
    def __init__(self, paths, labels, transform):
        self.paths     = paths
        self.labels    = labels
        self.transform = transform
    def __len__(self): return len(self.paths)
    def __getitem__(self, idx):
        img = cv2.imread(self.paths[idx], cv2.IMREAD_GRAYSCALE)
        return self.transform(img), self.labels[idx]

class ResNet50_Alzheimer(nn.Module):
    def __init__(self, num_classes=4):
        super().__init__()
        backbone       = models.resnet50(
            weights=models.ResNet50_Weights.IMAGENET1K_V2)
        backbone.conv1 = nn.Conv2d(
            1, 64, kernel_size=7, stride=2, padding=3, bias=False)
        in_feat        = backbone.fc.in_features
        backbone.fc    = nn.Identity()
        self.backbone  = backbone
        self.classifier = nn.Sequential(
            nn.Linear(in_feat, 512), nn.ReLU(), nn.Dropout(0.5),
            nn.Linear(512, 256),    nn.ReLU(), nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )
    def forward(self, x):
        return self.classifier(self.backbone(x))

def train_epoch(model, loader, criterion, optimizer, device):
    model.train()
    loss_sum, correct, total = 0, 0, 0
    for imgs, labels in loader:
        imgs, labels = imgs.to(device), labels.to(device)
        optimizer.zero_grad()
        out  = model(imgs)
        loss = criterion(out, labels)
        loss.backward(); optimizer.step()
        loss_sum += loss.item()
        correct  += (out.argmax(1)==labels).sum().item()
        total    += labels.size(0)
    return loss_sum/len(loader), correct/total*100

def validate_epoch(model, loader, criterion, device):
    model.eval()
    loss_sum, correct, total = 0, 0, 0
    all_preds, all_labels = [], []
    with torch.no_grad():
        for imgs, labels in loader:
            imgs, labels = imgs.to(device), labels.to(device)
            out          = model(imgs)
            loss_sum    += criterion(out, labels).item()
            preds        = out.argmax(1)
            correct     += (preds==labels).sum().item()
            total       += labels.size(0)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    acc = correct/total*100
    f1  = f1_score(all_labels, all_preds,
                   average='macro', zero_division=0)*100
    return loss_sum/len(loader), acc, f1, all_preds, all_labels

def train_kfold(image_paths, cog_labels,
                num_epochs=30, batch_size=32, lr=1e-4):
    device  = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    skf     = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    results = []

    for fold, (train_idx, val_idx) in enumerate(
            skf.split(image_paths, cog_labels)):

        print(f"\n{'='*55}\nFOLD {fold+1}/5\n{'='*55}")

        t_paths  = [image_paths[i] for i in train_idx]
        v_paths  = [image_paths[i] for i in val_idx]
        t_labels = cog_labels[train_idx]
        v_labels = cog_labels[val_idx]

        for i,n in enumerate(CLASS_NAMES):
            print(f"  Train {n:10s}: {(t_labels==i).sum()} "
                  f"| Val: {(v_labels==i).sum()}")

        train_ds = AlzheimerDataset(t_paths, t_labels, train_transform)
        val_ds   = AlzheimerDataset(v_paths, v_labels, val_transform)

        #  sqrt thay vì 1/count
        class_counts  = np.array([(t_labels==i).sum() for i in range(4)])
        sample_weight = np.array(
            [1.0/np.sqrt(class_counts[l]) for l in t_labels])
        sampler = WeightedRandomSampler(
            weights     = torch.tensor(sample_weight, dtype=torch.float),
            num_samples = len(t_labels),
            replacement = True)

        train_loader = DataLoader(train_ds, batch_size=batch_size,
                                  sampler=sampler,
                                  num_workers=2, pin_memory=True)
        val_loader   = DataLoader(val_ds, batch_size=batch_size,
                                  shuffle=False,
                                  num_workers=2, pin_memory=True)

        model = ResNet50_Alzheimer(num_classes=4).to(device)

        # class weight cũng dùng sqrt
        w = torch.tensor(
            [1.0/np.sqrt(class_counts[i]) for i in range(4)],
            dtype=torch.float32).to(device)
        # Normalize weights
        w = w / w.sum() * 4
        criterion = nn.CrossEntropyLoss(weight=w)

        optimizer = torch.optim.Adam([
            {'params': [p for n,p in model.named_parameters()
                        if 'classifier' not in n], 'lr': lr*0.1},
            {'params': model.classifier.parameters(), 'lr': lr}
        ], weight_decay=1e-4)
        scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
            optimizer, T_max=num_epochs, eta_min=1e-6)

        best_f1, best_acc = 0, 0
        history = {'train_loss':[], 'val_loss':[],
                   'train_acc': [], 'val_acc': [], 'val_f1':[]}

        for epoch in range(num_epochs):
            t_loss, t_acc = train_epoch(
                model, train_loader, criterion, optimizer, device)
            v_loss, v_acc, v_f1, v_preds, v_lbls = validate_epoch(
                model, val_loader, criterion, device)
            scheduler.step()

            history['train_loss'].append(t_loss)
            history['val_loss'].append(v_loss)
            history['train_acc'].append(t_acc)
            history['val_acc'].append(v_acc)
            history['val_f1'].append(v_f1)

            if v_f1 > best_f1:
                best_f1, best_acc = v_f1, v_acc
                best_preds, best_lbls_saved = v_preds, v_lbls
                torch.save(model.state_dict(),
                    f'{CHECKPOINT_DIR}/best_fold{fold+1}.pth')
                print(f"Epoch {epoch+1:3d} | "
                      f"Train: {t_acc:.1f}% | "
                      f"Val: {v_acc:.1f}% | "
                      f"F1: {v_f1:.1f}% ")
            else:
                print(f"Epoch {epoch+1:3d} | "
                      f"Train: {t_acc:.1f}% | "
                      f"Val: {v_acc:.1f}% | "
                      f"F1: {v_f1:.1f}%")

        # Plot + report
        plot_history(history, fold+1)
        plot_cm(best_lbls_saved, best_preds, fold+1)
        print(f"\nFold {fold+1} Best Report:")
        print(classification_report(best_lbls_saved, best_preds,
                                    target_names=CLASS_NAMES,
                                    zero_division=0))
        results.append({'fold':fold+1, 'acc':best_acc, 'f1':best_f1})

    # Tổng kết
    print(f"\n{'='*55}\nTỔNG KẾT 5-FOLD\n{'='*55}")
    accs = [r['acc'] for r in results]
    f1s  = [r['f1']  for r in results]
    for r in results:
        print(f"Fold {r['fold']}: Acc={r['acc']:.1f}% | F1={r['f1']:.1f}%")
    print(f"\nMean Acc: {np.mean(accs):.1f}% ± {np.std(accs):.1f}%")
    print(f"Mean F1 : {np.mean(f1s):.1f}% ± {np.std(f1s):.1f}%")
    return results

def plot_history(history, fold):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    ep = range(1, len(history['train_acc'])+1)
    ax1.plot(ep, history['train_loss'], label='Train')
    ax1.plot(ep, history['val_loss'],   label='Val')
    ax1.set_title(f'Fold {fold} - Loss')
    ax1.legend(); ax1.grid(True)
    ax2.plot(ep, history['train_acc'], label='Train Acc')
    ax2.plot(ep, history['val_acc'],   label='Val Acc')
    ax2.plot(ep, history['val_f1'],    label='Val F1', linestyle='--')
    ax2.set_title(f'Fold {fold} - Accuracy & F1')
    ax2.legend(); ax2.grid(True)
    plt.tight_layout()
    plt.savefig(f'{CHECKPOINT_DIR}/history_fold{fold}.png', dpi=100)
    plt.show()

def plot_cm(true_labels, pred_labels, fold):
    cm = confusion_matrix(true_labels, pred_labels)
    plt.figure(figsize=(7, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
    plt.title(f'Confusion Matrix - Fold {fold}')
    plt.ylabel('True'); plt.xlabel('Predicted')
    plt.tight_layout()
    plt.savefig(f'{CHECKPOINT_DIR}/cm_fold{fold}.png', dpi=100)
    plt.show()


image_paths, cog_labels = prepare_data(DATA_DIR)
results = train_kfold(
    image_paths, cog_labels,
    num_epochs = 30,
    batch_size = 32,
    lr         = 1e-4
)
```

```python
#  Load Fold 5 + kiểm tra baseline
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = ResNet50_Alzheimer(num_classes=4).to(device)
model.load_state_dict(torch.load(
    f'{CHECKPOINT_DIR}/best_fold5.pth', map_location=device))
print(" Loaded best_fold5.pth!")

# Tách 10% test set cố định
from sklearn.model_selection import train_test_split
image_paths, all_labels = prepare_data(DATA_DIR)

train_paths, test_paths, train_labels, test_labels = train_test_split(
    image_paths, all_labels,
    test_size=0.10, stratify=all_labels, random_state=42)

# DataLoader
class_counts  = np.array([(train_labels==i).sum() for i in range(4)])
sample_weight = np.array([1.0/np.sqrt(class_counts[l]) for l in train_labels])
sampler = WeightedRandomSampler(
    torch.tensor(sample_weight, dtype=torch.float),
    num_samples=len(train_labels), replacement=True)

train_ds = AlzheimerDataset(train_paths, train_labels, train_transform)
test_ds  = AlzheimerDataset(test_paths,  test_labels,  val_transform)

train_loader2 = DataLoader(train_ds, batch_size=32, sampler=sampler,
                           num_workers=2, pin_memory=True)
test_loader2  = DataLoader(test_ds,  batch_size=32, shuffle=False,
                           num_workers=2, pin_memory=True)

# Baseline
def evaluate(model, loader):
    model.eval()
    preds_all, lbls_all = [], []
    with torch.no_grad():
        for imgs, labels in loader:
            imgs  = imgs.to(device)
            preds = model(imgs).argmax(1).cpu().numpy()
            preds_all.extend(preds)
            lbls_all.extend(labels.numpy())
    acc = np.mean(np.array(preds_all)==np.array(lbls_all))*100
    f1  = f1_score(lbls_all, preds_all, average='macro', zero_division=0)*100
    return acc, f1, preds_all, lbls_all

acc0, f10, _, _ = evaluate(model, test_loader2)
print(f"Baseline → Acc: {acc0:.1f}% | F1: {f10:.1f}%")
```

```python
# Train thêm 20 epochs
from torch.optim.lr_scheduler import CosineAnnealingLR

EXTRA_EPOCHS = 20

w = torch.tensor([1.0/np.sqrt(class_counts[i]) for i in range(4)],
                  dtype=torch.float32).to(device)
w = w / w.sum() * 4
criterion = nn.CrossEntropyLoss(weight=w)

optimizer = torch.optim.Adam([
    {'params': [p for n,p in model.named_parameters()
                if 'classifier' not in n], 'lr': 1e-5},
    {'params': model.classifier.parameters(), 'lr': 1e-4}
], weight_decay=1e-4)
scheduler = CosineAnnealingLR(optimizer, T_max=EXTRA_EPOCHS, eta_min=1e-7)

best_f1, best_acc = f10, acc0
history = {'train_acc':[], 'test_acc':[], 'test_f1':[]}

print(f"TRAIN THÊM {EXTRA_EPOCHS} EPOCHS (từ F1={best_f1:.1f}%)\n")

for epoch in range(EXTRA_EPOCHS):
    model.train()
    correct, total = 0, 0
    for imgs, labels in train_loader2:
        imgs, labels = imgs.to(device), labels.to(device)
        optimizer.zero_grad()
        out  = model(imgs)
        loss = criterion(out, labels)
        loss.backward(); optimizer.step()
        correct += (out.argmax(1)==labels).sum().item()
        total   += labels.size(0)
    t_acc = correct/total*100

    v_acc, v_f1, v_preds, v_lbls = evaluate(model, test_loader2)
    scheduler.step()
    history['train_acc'].append(t_acc)
    history['test_acc'].append(v_acc)
    history['test_f1'].append(v_f1)

    if v_f1 > best_f1:
        best_f1, best_acc = v_f1, v_acc
        best_preds, best_lbls = v_preds, v_lbls
        torch.save(model.state_dict(),
                   f'{CHECKPOINT_DIR}/best_final.pth')
        print(f"Epoch {epoch+1:3d} | Train: {t_acc:.1f}% | "
              f"Test: {v_acc:.1f}% | F1: {v_f1:.1f}% SAVED")
    else:
        print(f"Epoch {epoch+1:3d} | Train: {t_acc:.1f}% | "
              f"Test: {v_acc:.1f}% | F1: {v_f1:.1f}%")

print(f"\n Best → Acc: {best_acc:.1f}% | F1: {best_f1:.1f}%")
```

```python
# Đánh giá cuối + Confusion Matrix
model.load_state_dict(torch.load(
    f'{CHECKPOINT_DIR}/best_final.pth', map_location=device))
_, _, final_preds, final_lbls = evaluate(model, test_loader2)

print("=" * 55)
print("KẾT QUẢ CUỐI - ResNet50 Kaggle 4 Classes")
print("=" * 55)
print(classification_report(final_lbls, final_preds,
                             target_names=CLASS_NAMES, zero_division=0))

cm = confusion_matrix(final_lbls, final_preds)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
plt.title(f'Confusion Matrix Final | F1={best_f1:.1f}%')
plt.ylabel('True'); plt.xlabel('Predicted')
plt.tight_layout()
plt.savefig(f'{CHECKPOINT_DIR}/cm_final.png', dpi=100)
plt.show()
```
